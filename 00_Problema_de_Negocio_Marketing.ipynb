# 📘 Problema de Negócio - Performance de Campanhas de Marketing

Este notebook ajuda a entender o problema antes de codar, definindo objetivos, métricas de sucesso e hipóteses.
Use-o como guia de entrevistas com stakeholders.

## Contexto do Negócio

**Objetivo:** melhorar a performance de campanhas de marketing, aumentando a taxa de conversão e otimizando custos.

**Impacto:** cada conversão gerada contribui para aumento de receita e redução de desperdício em campanhas ineficientes.
Definição de sucesso: cliente exposto à campanha que realizou a ação esperada (ex.: compra, inscrição, clique).

### Decisões-chave:
- **Janela de previsão:** prever a probabilidade de conversão após exposição à campanha.
- **População-alvo:** leads/clientes expostos às campanhas.
- **Estratégia operacional:** priorizar campanhas ou clientes com maior chance de conversão.
- **Métricas técnicas:** ROC-AUC, PR-AUC, Recall, Precision, Lift.
- **Métricas de negócio:** taxa de conversão, receita incremental, ROI da campanha.


### ROI (Retorno sobre Investimento)
**Receita incremental** = (Clientes convertidos atribuídos à campanha × Ticket médio)
**Custo da campanha** = investimento em mídia + custo operacional
**ROI** = (Receita incremental – Custo campanha) / Custo campanha

## Perguntas de Descoberta (para stakeholders)
- Como a empresa define conversão hoje (compra, clique, lead qualificado)?
- Qual o ticket médio e a margem média por produto/segmento?
- Quais canais e tipos de campanhas existem? (email, social, paid ads)
- Quais custos de campanha devem ser considerados (CPC, CPM, desconto)?
- Qual orçamento/capacidade operacional existe para priorização?
- Como será feita a integração dos insights no CRM/marketing automation?
- Como monitorar e atualizar a performance após o deploy?

## Métricas de Sucesso
**Técnicas:** ROC-AUC, PR-AUC, Recall, Lift.
**Negócio:** aumento da taxa de conversão, receita incremental, ROI líquido.

## 📂 Variáveis disponíveis (resumo)
**Empresa:** A empresa responsável pela campanha, representando uma mistura de marcas fictícias.

**Tipo_de_Campanha:** O tipo de campanha empregada, incluindo e-mail, mídia social, influenciador, display ou pesquisa.

**Público_Alvo:** O segmento específico de público-alvo da campanha, como mulheres de 25 a 34 anos, homens de 18 a 24 anos ou todas as faixas etárias.

**Duração:** A duração da campanha, expressa em dias.

**Canais_Usados:** Os canais utilizados para promover a campanha, que podem incluir e-mail, plataformas de mídia social, YouTube, sites ou Google Ads.

**Taxa_de_Conversão:** A porcentagem de leads ou impressões que se converteram em ações desejadas, indicando a eficácia da campanha.

**Custo_de_Aquisição:** O custo incorrido pela empresa para adquirir clientes, apresentado em formato monetário.

**ROI:** Retorno sobre o Investimento, representando a lucratividade e o sucesso da campanha.

**Localização:** A localização geográfica onde a campanha foi conduzida, abrangendo grandes cidades como Nova York, Los Angeles, Chicago, Houston ou Miami.

**Idioma:** O idioma usado na comunicação da campanha, incluindo inglês, espanhol, francês, alemão ou mandarim.

**Cliques:** o número de cliques gerados pela campanha, indicando o engajamento do usuário.

**Impressões:** o número total de vezes que a campanha foi exibida ou visualizada pelo público-alvo.

**Pontuação de Engajamento:** uma pontuação de 1 a 10 que mede o nível de engajamento gerado pela campanha.

**Segmento de Cliente:** o segmento específico de clientes ou categoria de público para o qual a campanha foi criada, como entusiastas de tecnologia, fashionistas, entusiastas de saúde e bem-estar, amantes da gastronomia ou aventureiros ao ar livre.

**Data:** a data em que a campanha foi realizada, fornecendo uma perspectiva cronológica para analisar tendências e padrões.
